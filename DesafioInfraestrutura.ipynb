```python

from Extraction.Extraction_Data import extract_data

from Transform.Transform_basic_infra import calculate_basic_infra_percent

from Transform.Transform_advanced_infra import calculate_advanced_infra_percent

from Load.Load_basic_infra import visualize_basic_infra

from Load.Load_advanced_infra import visualize_advanced_infra


def main():
    df_paraiba = extract_data("Tabelas/microdados_ed_basica_2024.csv")

    print(f"Total de escolas públicas na Paraíba e em funcionamento { len(df_paraiba) }")

    infra_basicas = {
        'IN_AGUA_POTAVEL': 'Água potável',
        'IN_AGUA_REDE_PUBLICA': 'Água da rede pública',
        'IN_ENERGIA_REDE_PUBLICA': 'Energia da rede pública',
        'IN_ESGOTO_REDE_PUBLICA': 'Esgoto da rede pública',
        'IN_BANHEIRO': 'Banheiro',
        'IN_BIBLIOTECA': 'Biblioteca',
        'IN_INTERNET': 'Internet',
        'IN_ALIMENTACAO': 'Alimentação escolar',
        'IN_REFEITORIO': "Refeitorio escolar"
    }
    
    infra_avancada = {
        'IN_AGUA_POTAVEL': 'Água potável',
        'IN_AGUA_REDE_PUBLICA': 'Água da rede pública',
        'IN_ENERGIA_REDE_PUBLICA': 'Energia da rede pública',
        'IN_ESGOTO_REDE_PUBLICA': 'Esgoto da rede pública',
        'IN_BANHEIRO': 'Banheiro',
        'IN_BIBLIOTECA': 'Biblioteca',
        'IN_INTERNET': 'Internet',
        'IN_ALIMENTACAO': 'Alimentação escolar',
        'IN_REFEITORIO': "Refeitorio escolar",
        'IN_ALMOXARIFADO':"Almoxarifado escolar",
        'IN_LABORATORIO_INFORMATICA': 'Laboratório de informática',
        'IN_QUADRA_ESPORTES': 'Quadra de esportes',
        'IN_SALA_DIRETORIA': 'Sala da diretoria',
        'IN_SALA_PROFESSOR': 'Sala dos professores',
        'IN_SECRETARIA': 'Sala da secretaria',
        'IN_EQUIP_IMPRESSORA': 'Impressora',
        'IN_COMPUTADOR': 'Computador'
    }
    
    df_porcentagem_infra_basica_faltando = calculate_basic_infra_percent(df_paraiba,    infra_basicas)

    print("\nPorcentagem de componente básico faltando nas escolas")

    print(df_porcentagem_infra_basica_faltando)

    visualize_basic_infra(df_porcentagem_infra_basica_faltando)

    df_porcentagem_infra_avancada_constando = calculate_advanced_infra_percent(df_paraiba,infra_avancada)

    print("\nPorcentagem de componentes avançados/basicos presentes nas escolas")

    print(df_porcentagem_infra_avancada_constando)

    visualize_advanced_infra(df_porcentagem_infra_avancada_constando)

main()

```
